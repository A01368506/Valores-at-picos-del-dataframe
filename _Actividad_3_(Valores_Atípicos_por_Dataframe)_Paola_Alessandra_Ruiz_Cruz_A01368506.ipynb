```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
gastos_costos = pd.ExcelFile('/content/Gastos y costos 20-23.xlsx')
hoja = '2023'
g_c= gastos_costos.parse(hoja)

g_c
```

```python
g_c.info()
```

```python
g_c.head()
```

```python
df = g_c.iloc[2:]
```

```python
df
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df
```

```python
df.info()
```

```python
valores_nulos = df.isnull().sum()
valores_nulos
```

```python
data1 = df.fillna(method ="bfill")
data1
```

```python
valores_nulos = data1.isnull().sum()
valores_nulos
```

```python
data1["Poliza"]= data1["Poliza"].fillna("PE03-ENE23")
data1

```

```python
valores_nulos2 = data1.isnull().sum()
valores_nulos2
```

```python
data1.to_csv("NAS.csv")
from google.colab import files


files.download("NAS.csv")
```

```python
data2 = data1.drop("Otros", axis=1)
data2
```

```python
valores_nulos3 = data2.isnull().sum()
valores_nulos3
```

```python
data2["MP"] =data2["MP"].fillna(method="ffill")
data2
```

```python
data2["FP"] =data2["FP"].fillna(method="ffill")
```

```python
valores_nulos4 = data2.isnull().sum()
valores_nulos4
```

```python
data2
```

```python
fig = plt.figure(figsize=(15,8))
data2.plot(kind = "box", vert =False )
plt.title("Valores atipicos del dataframe")
plt.show()
```

```python
y = data2
Limite_superior= y.mean() + 3*y.std()
Limite_inferior= y.mean() - 3*y.std()
print("Limite superior permitido: ",Limite_superior)
print("Limite inferior permitido: ",Limite_inferior)
```

```python
data2.to_csv("DSV.csv")
from google.colab import files


files.download("DSV.csv")
```

```python
data3 = data2[(y<=Limite_superior)&(y>=Limite_inferior)]
data3
```

```python
data3.to_csv("ATIP.csv")
from google.colab import files


files.download("ATIP.csv")
```
